```python
from copy import deepcopy

import numpy as np

from openscm.units import unit_registry
from openscm.models import PH99Model

import matplotlib.pyplot as plt
plt.style.use('seaborn-dark')
```

```python
ph99 = PH99Model()
```

```python
ph99.emissions = np.sin(np.linspace(0, 100, 100) * 2 * np.pi / 40) * unit_registry("GtC/yr")
# I need to add a setter which sets other arrays based on length of emissions
# and does re-setting etc.
# temporary workaround
initialiser = np.nan * np.zeros_like(ph99.emissions)

cumulative_emissions_init = deepcopy(initialiser)
cumulative_emissions_init[0] = 0
ph99.cumulative_emissions = unit_registry.Quantity(
    cumulative_emissions_init,
    "GtC"
)

concentrations_init = deepcopy(initialiser)
concentrations_init[0] = 291
ph99.concentrations = unit_registry.Quantity(
    concentrations_init,
    "ppm"
)

temperatures_init = deepcopy(initialiser)
temperatures_init[0] = 14.6
ph99.temperatures = unit_registry.Quantity(
    temperatures_init,
    "degC"
)
```
